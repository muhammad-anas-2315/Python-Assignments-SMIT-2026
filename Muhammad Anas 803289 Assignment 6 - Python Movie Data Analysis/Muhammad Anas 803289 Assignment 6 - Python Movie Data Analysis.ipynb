### **Question 36**
(Visualization) 'Drama' aur 'Action' genres ki movies ke 'Gross' revenue ko compare karne ke liye ek Grouped Box Plot ya Violin Plot banayein.

```python
df_exploded = movie_data.copy()
df_exploded['Genre'] = df_exploded['Genre'].str.split(',').str.strip()
df_exploded = df_exploded.explode('Genre')
compare_df = df_exploded[df_exploded['Genre'].isin(['Action','Drama'])]
compare_df['Gross_Millions'] = compare_df['Gross'] / 1_000_000

plt.figure(figsize=(9,6))
sns.set_theme(style='whitegrid')
sns.violinplot(data=compare_df, x='Genre', y='Gross_Millions', hue='Genre', palette='muted',
              inner='quartile', legend=False)
plt.title('Gross Revenue Comparison: Drama vs Action',fontsize=16, fontweight='bold', pad = 15)
plt.xlabel('Movie Genre',fontsize=12)
plt.ylabel('Gross Revenue',fontsize=12)
plt.show()
```

### **Question 40**
(Visualization) Top 5 highest rated movies aur unke 'Meta_score' vs 'IMDB_Rating' (scaled to 100) ka ek Side-by-Side Comparison Bar Chart banayein.

```python
top_5_hi_rated_movies = movie_data.sort_values(by='IMDB_Rating',ascending=False).head(5).copy()
top_5_hi_rated_movies['IMDB_Rating_Scaled'] = top_5_hi_rated_movies['IMDB_Rating']*10
melted_df = pd.melt(top_5_hi_rated_movies, id_vars=['Series_Title'],
    value_vars=['Meta_score', 'IMDB_Rating_Scaled'],
    var_name='Rating_Type',
    value_name= 'Score',
)
plt.figure(figsize=(12,6))
sns.set_theme(style='whitegrid')
sns.barplot(data=melted_df, x='Series_Title', y= 'Score', hue='Rating_Type',
            palette='Set2')
plt.title('Top 5 Movies: Meta Score vs IMDB Rating (Scaled to 100)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Movie Title', fontsize=12, labelpad= 10)
plt.ylabel('Score (Out of 100)', fontsize=12,)
plt.ylim(0,110)
plt.xticks(rotation=15, ha= 'right')
plt.legend(title='Rating Metric')
plt.tight_layout()
plt.show()
```
