#### Q10: Count Vowels in String

```python
text = """Data science is an interdisciplinary field that combines statistics, mathematics,
computer science, and business expertise to extract actionable insights from structured and unstructured data."""
vowels = "aeiou"
count = 0
for ch in text.lower():
    if ch in vowels:
        count += 1
print(count)
```
