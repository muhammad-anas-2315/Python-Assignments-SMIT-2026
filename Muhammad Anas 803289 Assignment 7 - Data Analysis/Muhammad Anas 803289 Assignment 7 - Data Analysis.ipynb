```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Dataset 1 (Loan Dataset):

```python
loan_train_data = pd.read_csv('loan-train.csv')
```

```python
loan_train_data.shape
```

```python
loan_train_data.info()
```

```python
loan_train_data.isna().sum()
```

```python
loan_train_data.head()
```

```python
loan_train_data.duplicated().sum()
```

```python
loan_train_data[loan_train_data['Gender'].isna()]
```

```python
loan_train_data['Gender'] = loan_train_data['Gender'].fillna('Unknown')
```

```python
loan_train_data.info()
```

```python
loan_train_data[loan_train_data['Married'].isna()]
```

```python
loan_train_data['Married'] = loan_train_data['Married'].fillna('Unspecified')
```

```python
loan_train_data.info()
```

```python
loan_train_data[loan_train_data['Dependents'].isna()]
```

```python
loan_train_data['Dependents'] = loan_train_data['Dependents'].fillna(0)
```

```python
loan_train_data.info()
```

```python
loan_train_data[loan_train_data['Self_Employed'].isna()]
```

```python
loan_train_data['Self_Employed'] = loan_train_data['Self_Employed'].fillna('No')
```

```python
loan_train_data.info()
```

```python
loan_train_data[loan_train_data['LoanAmount'].isna()]
```

```python
loan_train_data['LoanAmount']
```

```python
sns.boxplot(loan_train_data['LoanAmount'])
```

```python
q1 = loan_train_data['LoanAmount'].quantile(0.25)
q3 = loan_train_data['LoanAmount'].quantile(0.75)
```

```python
iqr = q3 - q1
low_limit = q1 - 1.5 * iqr
upp_limit = q3 + 1.5 * iqr
print(low_limit)
print(upp_limit)
```

```python
loan_train_data = loan_train_data[(loan_train_data['LoanAmount'] >= low_limit) & (loan_train_data['LoanAmount'] <= upp_limit)]
```

```python
loan_train_data.info()
```

```python
loan_train_data.head()
```

```python
loan_train_data[loan_train_data['Loan_Amount_Term'].isna()]
```

```python
loan_train_data['Loan_Amount_Term'].value_counts()
```

```python
loan_train_data['Loan_Amount_Term'] = loan_train_data['Loan_Amount_Term'].fillna(loan_train_data['Loan_Amount_Term'].mode()[0])
```

```python
loan_train_data.info()
```

```python
loan_train_data[loan_train_data['Credit_History'].isna()]
```

```python
loan_train_data['Credit_History'] = loan_train_data['Credit_History'].fillna(loan_train_data['Loan_Status'].map({'Y':1.0, 'N':0.0}))
```

```python
loan_train_data.info()
```

```python
loan_train_data.to_csv('loan-train_Analysis.csv')
```

#### Dataset 2 (Adult Dataset):

```python
adult_dataset = pd.read_csv('adult.csv')
```

```python
adult_dataset.head()
```

```python
adult_dataset.isna().sum()
```

```python
adult_dataset.info()
```

```python
adult_dataset[adult_dataset[['workclass','occupation','native-country']].isin(['?','NaN']).any(axis=1)]
```

```python
adult_dataset.replace('?',np.nan, inplace=True)
```

```python
adult_dataset.duplicated().sum()
```

```python
adult_dataset[adult_dataset.duplicated()]
```

```python
adult_dataset.drop_duplicates(inplace=True)
```

```python
adult_dataset.info()
```

```python
adult_dataset['workclass'] = adult_dataset['workclass'].fillna(adult_dataset['workclass'].mode()[0])
```

```python
adult_dataset.info()
```

```python
adult_dataset['occupation'].value_counts()
```

```python
adult_dataset['occupation'].isna().sum()
```

```python
adult_dataset['occupation'] = adult_dataset['occupation'].fillna(adult_dataset['occupation'].mode()[0])
```

```python
adult_dataset.info()
```

```python
adult_dataset['native-country'] = adult_dataset['native-country'].fillna(adult_dataset['native-country'].mode()[0])
```

```python
adult_dataset.info()
```

```python
adult_dataset.to_csv('adult_dataset_analysis.csv')
```

```python
sns.countplot(x='income', data=adult_dataset, palette='Set2')
plt.title('Income Distribution')
plt.show()
```

```python
sns.boxplot(x='income', y='age', data=adult_dataset)
plt.title('Age Distribution by Income Level')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='education', hue='income', data=adult_dataset, palette='viridis')
plt.xticks(rotation=45)
plt.title('Income Level by Education')
plt.tight_layout()
plt.show()
```

```python
sns.violinplot(x='income', y='hours-per-week', data=adult_dataset, inner='quartile')
plt.title('Work Hours per Week vs Income')
plt.show()
```

```python
adult_dataset['income_encoded'] = adult_dataset['income'].apply(lambda x: 1 if x == '>50K' else 0)

numerical_df = adult_dataset.select_dtypes(include=[np.number])
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
hotel_book_dataset = pd.read_csv('hotel_booking.csv')
```

```python
hotel_book_dataset.info()
```

```python
hotel_book_dataset.isna().sum()
```

```python
hotel_book_dataset.duplicated().sum()
```

```python
hotel_book_dataset[hotel_book_dataset['country'].isna()]
```

```python
hotel_book_dataset[hotel_book_dataset['children'].isna()]
```

```python
hotel_book_dataset['country'] = hotel_book_dataset['country'].fillna(hotel_book_dataset['country'].mode()[0])
```

```python
hotel_book_dataset['company'] = hotel_book_dataset['company'].fillna(hotel_book_dataset['company'].mode()[0])
```

```python
hotel_book_dataset['agent'] = hotel_book_dataset['agent'].fillna(hotel_book_dataset['agent'].mode()[0])
```

```python
hotel_book_dataset['children'] = hotel_book_dataset['children'].fillna(hotel_book_dataset['children'].mode()[0])
```

```python
hotel_book_dataset.isna().sum()
```

```python
hotel_book_dataset.info()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='is_canceled', data=hotel_book_dataset, palette='Set2')
plt.title('Booking Cancelation Status (0 = Not Canceled, 1 = Canceled)')
plt.xlabel('Is Canceled')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='hotel', hue='is_canceled', data=hotel_book_dataset, palette='viridis')
plt.title('Cancelation Trends by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.legend(title='Is Canceled', labels=['No', 'Yes'])
plt.show()
```

```python
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 
                'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(12, 6))
sns.countplot(x='arrival_date_month', data=hotel_book_dataset, order=months_order, palette='pastel')
plt.title('Busy Months for Hotel Bookings')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(hotel_book_dataset['lead_time'], bins=50, kde=True, color='purple')
plt.title('Distribution of Lead Time')
plt.xlabel('Lead Time (Days)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='market_segment', hue='is_canceled', data=hotel_book_dataset, palette='magma')
plt.title('Bookings and Cancelations by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Canceled', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(x='reserved_room_type', data=hotel_book_dataset, order=sorted(hotel_book_dataset['reserved_room_type'].unique()), color='skyblue', label='Reserved')
sns.countplot(x='assigned_room_type', data=hotel_book_dataset, order=sorted(hotel_book_dataset['assigned_room_type'].unique()), color='orange', alpha=0.6, label='Assigned')
plt.title('Comparison of Reserved vs Assigned Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.countplot(ax=axes[0], x='customer_type', data=hotel_book_dataset, palette='Set3')
axes[0].set_title('Bookings by Customer Type')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=30)

sns.countplot(ax=axes[1], x='deposit_type', data=hotel_book_dataset, palette='Set2')
axes[1].set_title('Bookings by Deposit Type')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='is_repeated_guest', y='total_of_special_requests', data=hotel_book_dataset, palette='spring')
plt.title('Special Requests: New vs Repeated Guests')
plt.xlabel('Is Repeated Guest (0 = No, 1 = Yes)')
plt.ylabel('Number of Special Requests')
plt.show()
```
