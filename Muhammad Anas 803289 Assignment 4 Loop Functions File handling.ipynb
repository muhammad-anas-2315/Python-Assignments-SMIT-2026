**Q11: Ek list ke numbers ka square nikal kar nayi list banayein.**

**Q12: Check karein ke number Prime hai ya nahi.**

```python
num = int(input("Enter a Number: "))
if num >1:
    is_prime = True
    for i in range (2,num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(f"{num} ek prime number hai")
    else:
        print(f"{num} ek prime number nahi hai")
    
```

**Q16: Factorial find karein (e.g., 5!).**

```python
num = 5
factorial = 1
if num < 0:
    print("Negative Number Factorial ni nikalta")
elif num == 0:
    print("0 ka Factorial 1 hota hai.")
else:
    for i in range(1,num+1):
        factorial *=i
    print(f"{num} ka Factorial Hai: {factorial}")
```

**Q20: Number Guessing Game (3 chances).**

```python
import random
secret_num = random.randint(1,10)
chances = 3
attempt = 0
guess = ""
print("---Number Guessing Game---")
print("Total Chances",chances)
while guess != secret_num:
    guess = int(input("Enter Guess a Number (3 Chances): "))
    attempt +=1
    if guess == secret_num:
        print(f"Congratulations Apne {attempt} me Correct Number Guess Kia.")
        break
    elif attempt == 3:
        break
    elif guess < secret_num:
        print("Small Number Entered")
    elif guess > secret_num:
        print("Large Number Entered")
print(f"\n Game Over Attempts: {attempt}")
```

**Q23: Arbitrary arguments (`*args`) ka use karke function banayein jo saare numbers ko multiply kare.**

```python
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result
print(multiply(5,9))
print(multiply(9,8,2))
```

**Q24: Keyword arguments (`**kwargs`) ka use karke student ki info print karein (Name, Age).**

```python
def student_info(**kwargs):
    print("---Student Information ---")
    for key, value in kwargs.items():
        print(f"{key.title()}: {value}")
student_info(name="Anas", age=24, city="Karachi", course="Python")
```

**Q27: Function banayein jo list accept kare aur uska average return kare.**

```python
def calculate_avg(num_list):
    if len(num_list) == 0:
        return 0
    total_sum = sum(num_list)
    total_count = len(num_list)
    average = total_sum / total_count
    return average
my_num = {10,20,30,40,50}
result = calculate_avg(my_num)
print(result)
```

#### **Q29: Recursive function banayein jo factorial calculate kare.**

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
print(factorial(4))
```

#### **Q30: Function banayein jo user se unlimited numbers le (*args) aur sirf even numbers ki list return kare.**

```python
def get_even_num(*args):
    even_list = []
    for num in args:
        if num % 2 ==0:
            even_list.append(num)
    return even_list
result = get_even_num (1,2,3,4,5,6,7,8,9)
print(result)
```
