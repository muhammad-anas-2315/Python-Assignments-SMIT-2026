**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
class Bank_Account:
    def __init__(self, account_holder, initial_balance=0):
        self.account_holder = account_holder

        self.__balance = initial_balance

    def get_balance(self):
        return self.__balance
    
    def Deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Rs. {amount} Deposited!")
        else:
            print("Error Invalid Amount")

    def withdraw(self, amount):
        if amount <=0:
            print("Error amount is invalid")
        elif amount > self.__balance:
            print(f"Insufficient Balance! Your current balance is: Rs. {self.__balance}")
        else:
            self.__balance -= amount
            print("Transaction Successful")    
```

**Q7:** Multilevel Inheritance ka use karein. Class `Animal` -> Class `Mammal` -> Class `Dog`. Dog class ke object se parent classes ke features ko access karein.

```python
class Animal:
    def __init__(self, kingdom):
        self.kingdom = kingdom

    def eat(self):
        return "Animal is Eating."

class Mammal(Animal):
    def __init__(self, kingdom, has_fur):
        super().__init__(kingdom)
        self.has_fur = has_fur

    def walk(self):
        return "Mammal is walking"

class Dog(Mammal):
    def __init__(self, kingdom, has_fur,breed):
        super().__init__(kingdom, has_fur)
        self.breed = breed
        
    def bark(self):
        return "Dog is bark!"
```

```python
my_dog = Dog("Animalia",True,"German Shepherd")
print(f'''---Attributes
(Kingdom) Animal Class: {my_dog.kingdom}
(Mammal) Has_fur: {my_dog.has_fur}
(Dog) Brees: {my_dog.breed}

--- Methods ---
{my_dog.eat()}
{my_dog.walk()}
{my_dog.bark()}
''')
```

**Q8:** Multiple Inheritance ka istemal karte hue do parent classes `Father` aur `Mother` banayein, aur unse aik child class `Child` derive karein jo dono ke habits inherit kare.

```python
class Father:
    def __init__(self):
        self.father_habit = "Wake up Early!"
    def gardening(self):
        return "My Father Like Gardening."

class Mother:
    def __init__(self):
        self.mother_habit = "Books Reading"
    def cooking(self):
        return "Like Cooking"
class Child(Father, Mother):
    def __init__(self):
        Father.__init__(self)
        Mother.__init__(self)
        self.child_habit = "Video Gamming"
  
    def display_habits(self):
        print(f'''
        Father Habits: {self.father_habit}
        Mother Habits: {self.mother_habit}
        Child Habits: {self.child_habit}
        ''')
```

```python
child = Child()
child.display_habits()
print(child.gardening())
print(child.cooking())
```

**Q10:** Hierarchical Inheritance kya hoti hai? Aik parent class `Shape` banayein aur us se do alag child classes `Circle` aur `Square` banakar samjhayein.

```python
class Shape:
    def __init__(self, color):
        self.color = color
    def drawing(self):
        return f"{self.color} Color Shape Drawing..."

class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius
    def area_circle(self):
        return 3.14 * (self.radius ** 2)

class Square(Shape):
    def __init__(self, color, side):
        super().__init__(color)
        self.side = side
    def area_square(self):
        return self.side * self.side
```

```python
my_circle = Circle(color="red", radius=5)
print(my_circle.drawing())
print(my_circle.area_circle())
print("-" * 25)
my_square = Square(color="Blue",side=4)
print(my_square.drawing())
print(my_square.area_square())
```

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages
    
    def __add__(self, other):
        total_pages = self.pages + other.pages
        return total_pages
```

```python
b1 = Book("Python Basics", 300)
b2 = Book("Oops Advanced", 450)
total = b1 + b2
print(total)
```

```python
class My_String:
    def __init__(self, text):
        self.text = text
    def __len__(self):
        count = 0
        for chracter in self.text:
            count += 1
        return count

class My_List:
    def __init__(self, item_list):
        self.item_list = item_list
    def __len__(self):
        count = 0
        for item in self.item_list:
            count += 1
        return count
class My_Dict:
    def __init__(self, dict_data):
        self.dict_data = dict_data
    def __len__(self):
        count = 0
        for key in self.dict_data:
            count += 1
        return count
```

**Q17:** Q16 wali `Computer` class ko inherit karke do child classes `Laptop` aur `Desktop` banayein aur dono mein `process()` method ko mandatory implement karein.

```python
from abc import ABC, abstractmethod

class Computer(ABC):
    @abstractmethod
    def process(self):
        pass

class Laptop(Computer):
    def __init__(self, brand):
        self.brand = brand
    
    def process(self):
        return f"{self.brand} Laptop Process in Portable Battery!"

class Desktop(Computer):
    def __init__(self, processor):
        self.processor = processor
    
    def process(self):
        return f"{self.processor} Desktop Process in Direct Power Supply."

my_laptop = Laptop("Dell")
print(my_laptop.process())
print(my_laptop.brand)

my_desktop = Desktop("Intel i7")
print(my_desktop.process())
print(my_desktop.processor)
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
from abc import ABC, abstractmethod

class Car(ABC):
    def start_engine(self):
        return "Engine Started... Vroom! Vroom!"
    
    @abstractmethod
    def drive(self):
        pass

class Sports_Car(Car):
    def __init__(self, name):
        self.name = name
    def drive(self):
        return f"{self.name} Now 200 km/h Speed!"

my_car = Sports_Car("Ferrari")
print(my_car.name)
print(my_car.start_engine())
print(my_car.drive())
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    def __init__(self, amount):
        self.amount = amount

    def print_receipt(self):
        print(f"Amount: {self.amount} Successfully Processed.")
    
    @abstractmethod
    def process_payment(self):
        pass

class CreditCardPayment(Payment):
    def __init__(self, amount, card_number, cvv):
        super().__init__(amount)
        self.card_number = card_number
        self.cvv = cvv
    def process_payment(self):
        print(f"--- Credit Card Processing ---")
        print(f"Connecting to Bank Server...")
        print(f"Verifying Card: XXXX-XXXX-XXXX-{self.card_number[-4:]}")
        print(f"Authenticating CVV...")
        print(f"Rs.{self.amount} debited via Dedit Card.")

class UPIPayment(Payment):
    def __init__(self, amount, upi_id):
        super().__init__(amount)
        self.upi_id = upi_id
    def process_payment(self):
        print(f"--- UPI / Mobile Wallet Processing ---")
        print(f"Sending request to UPI ID: {self.upi_id}")
        print(f"Waiting for user to enter 4-digit MPIN on phone...")
        print(f"MPIN Verified! Rs.{self.amount} transferred instantly.")

payment_01 = CreditCardPayment(amount=5000, card_number= "12345678001232012", cvv= 999)
payment_01.process_payment()
print("-"*20)
payment_01.print_receipt()

print("-"*20)
print("-"*20)

payment_02 = UPIPayment(amount=1200, upi_id="anas@upi")
payment_02.process_payment()
payment_02.print_receipt()
```
